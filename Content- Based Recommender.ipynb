###  Genres based recommendation

```python
#In this recommendation, movies.csv file has been used. Because df frame is created memory error because of being too big
```

```python
movies['genres'] = movies['genres'].str.split('|')
movies['genres'] = movies['genres'].fillna("").astype('str') # Convert genres to string value
movies['genres'] = movies['genres'].str.lower()
movies.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#Vectorization is the process of executing operations on entire arrays.
#Convert a collection of text documents to a matrix of token counts
#http://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html
vectorizer = CountVectorizer(analyzer='word',ngram_range=(1, 4),min_df=0, stop_words='english')
# accurancy scores for different n_ranges --> (1,4) ==  0.092692
#                                             (1,3) ==  0.091664 
#                                             (1,2) ==  0.087353
#                                             (1,1) ==  0.066824

y=movies['title']
X= vectorizer.fit_transform(movies['genres'])
X.shape
```

```python
vectorizer.get_feature_names()
```

#### Multinomial Naive Bayes
###### Feature vectors represent the frequencies with which certain events have been generated by a multinomial distribution. This is the event model typically used for document classification.

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
import numpy as np
```

```python
classifier = MultinomialNB()
classifier.fit(X,y)
```

```python
y_pred = classifier.predict(X)
metrics.accuracy_score(y,y_pred)
```

```python
actual = np.array(y)
indices = pd.Series(movies.index, index=movies['title'])
def movie(title):
    idx = indices[title]
    return classifier.predict(X[idx])
```

###  Rating based recommendation

```python
rating_of_movies = pd.DataFrame((df.groupby('title'))['rating'].mean())
rating_of_movies['count'] = pd.DataFrame(df.groupby('title')['rating'].count())
rating_of_movies.sort_values('count', ascending= False).head(4)
```

```python
#mov['title'] = df
```

```python
rating_of_movies.info()
```

```python
movies.isnull().sum()
```

```python
vote_averages = rating_of_movies['rating']
vote_counts = rating_of_movies['count']
C = vote_averages.mean()
C
```

```python
m = vote_counts.quantile(0.60)
m
```

```python
def wr(title):
    return (vote_counts[title] / (vote_counts[title] + m)* vote_averages[title]) + (m/(vote_averages[title] + m)*C)
```

```python
wr('Toy Story (1995)')
```
